WINE 데이터 세트(이진분류)

https://archive.ics.uci.edu/dataset/186/wine+quality


```python
import pandas as pd
import numpy as np

wine = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/MachineLearning/data/wine+quality/winequality.csv')
wine = wine.fillna(0)   # nan을 0으로 채움
print(wine)
```

```python
wine.head()
```

```python
wine.info()
```

```python
wine.describe()
```

```python
# 데이터프레임을 넘파이로 바꿈.
data = wine[['alcohol', 'residual sugar', 'pH']].to_numpy()

# class = 0이면 레드와인, 1이면 화이트와인
target = wine['class'].to_numpy()
```

```python
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    data, target, test_size=0.2, random_state=42)  # stratify=target :타겟에 있는 분포에 따라 분류하겠다
```

```python
print(train_input.shape, test_input.shape)
```

```python
LABELS = ["Red_Wine", "White_Wine"]

import matplotlib.pyplot as plt
count_classes = pd.value_counts(wine['class'], sort=True)
count_classes.plot(kind='bar', rot=0)
plt.title("Wine distribution")
plt.xticks(range(2), LABELS)
plt.xlabel("Wine class")
plt.ylabel("Frequency")
```

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

# 정규화
ss.fit(train_input)

train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)

print(train_scaled)
```

##1. 결정 트리


```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)
dt.fit(train_scaled, train_target)

print(dt.score(train_scaled, train_target))
print(dt.score(test_scaled, test_target))
#과대적합
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(10,7))
plot_tree(dt)
plt.show()
```

```python
# 위의 트리가 너무 복잡해서 간략하게 줄임
plt.figure(figsize=(10,7))
plot_tree(dt, max_depth=1, filled=True, feature_names=['alcohol', 'residual sugar', 'pH'])
plt.show()

# 조건을 만족하면 왼쪽, 아니면 오른쪽
# samples : 총 샘플 수
# 양성이 커질수록 파란색, 음성이 커질수록 주황색
```

##2. 랜덤 포레스트

```python
from sklearn.model_selection import cross_validate
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(rf, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
rf.fit(train_input, train_target)
print(rf.feature_importances_)
```

```python
rf = RandomForestClassifier(oob_score=True, n_jobs=-1, random_state=42)

rf.fit(train_input, train_target)
print(rf.oob_score_)
```

##3. 엑스트라 트리

```python
from sklearn.ensemble import ExtraTreesClassifier

et = ExtraTreesClassifier(n_jobs=-1, random_state=42)
scores = cross_validate(et, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
et.fit(train_input, train_target)
print(et.feature_importances_)
```

##4. 그래디언트 부스팅

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))

#과대적합에 강함
```

```python
gb = GradientBoostingClassifier(n_estimators=500, learning_rate=0.2, random_state=42)
scores = cross_validate(gb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))
```

```python
gb.fit(train_input, train_target)
print(gb.feature_importances_)
```

##5. 히스토그램 기반 그래디언트 부스팅


```python
from sklearn.ensemble import HistGradientBoostingClassifier

hgb = HistGradientBoostingClassifier(random_state=42)
scores = cross_validate(hgb, train_input, train_target, return_train_score=True, n_jobs=-1)

print(np.mean(scores['train_score']), np.mean(scores['test_score']))

#과대적합을 억제하면서 그래디언트 부스팅보다 높은 성능을 제공
```

```python
from sklearn.inspection import permutation_importance

hgb.fit(train_input, train_target)
# 특성을 하나씩 랜덤하게 섞어서 모델의 성능 변화를 관찰하여 어떤 특성이 중요한지 계산
result = permutation_importance(hgb, train_input, train_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
result = permutation_importance(hgb, test_input, test_target, n_repeats=10,
                                random_state=42, n_jobs=-1)
print(result.importances_mean)
```

```python
hgb.score(test_input, test_target)
```
